```python
#Import Libraries
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv')
df.head()
```

```python
#Get the number of unique pokemon
df['Name'].nunique()

# Get the names of Pokemon in alphabetical order
sorted_names = df['Name'].sort_values().unique()

# Display the first few names
print("Pokemon names in alphabetical order:")
print(sorted_names[:10])  # Displaying first 10 names as an example

# If you want to see all names, use:
# print(sorted_names)
```

```python
correlation_matrix = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr()

print(correlation_matrix)
```

```python
#plotting the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Correlation Heatmap")

#theme
sns.set_theme(style="whitegrid")

sns.heatmap(correlation_matrix, annot=True, cmap="viridis")

# Display the plot
plt.show()
```
